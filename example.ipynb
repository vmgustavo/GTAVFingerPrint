```python
import numpy as np
import requests
from PIL import Image
from io import BytesIO
```

```python
from FPMatch import FPMatch
```

```python
examples = [
    'https://i.imgur.com/fpoicQp.jpg',
    'https://i.imgur.com/kKXnTHt.png',
    'https://i.imgur.com/mZP69Z7.png',
    'https://i.imgur.com/zvQmwB0.png'
]

def request_image(url):
    r = requests.get(url)
    i = Image.open(BytesIO(r.content))
    return i

images = [np.array(request_image(url)) for url in examples]
```

```python
for image in images:
    FPMatch(np.array(image))
```
